#### 33. How to get the dates of yesterday, today and tomorrow? (★☆☆)

````python
# 1. 获取今天 (注意：'today' 会根据当前系统时间生成日期)
today = np.datetime64("today", "D")

# 2. 计算昨天 (今天减去 1 天)
yesterday = today - np.timedelta64(1, "D")

# 3. 计算明天 (今天加上 1 天)
tomorrow = today + np.timedelta64(1, "D")

print("昨天: ", yesterday)
print("今天: ", today)
print("明天: ", tomorrow)
````

#### 38. Consider a generator function that generates 10 integers and use it to build an array (★☆☆)

````python
def generate_integers():
    for i in range(10):
        yield i  # 每次只“吐”出一个数字，而不是一次性生成一个大列表


# 第二步：使用 np.fromiter 将生成器转化为 NumPy 数组
# 必须指定 dtype，这是为了让 NumPy 在底层直接分配连续的内存
Z = np.fromiter(generate_integers(), dtype=float, count=-1)

print(Z)
````

#### 44. Consider a random 10x2 matrix representing cartesian coordinates, convert them to polar coordinates (★★☆)

````python
matrix=np.random.random((10,2))
X, Y = matrix[:, 0], matrix[:, 1]
r=np.sqrt(X**2 + Y**2)
t=np.arctan2(Y, X)
ans=np.c_[r, t]
print(ans)
````

#### 48. Print the minimum and maximum representable values for each numpy scalar type (★★☆)

````python
for dtype in [np.int8, np.int32, np.int64]:
    print(np.iinfo(dtype).min)
    print(np.iinfo(dtype).max)
for dtype in [np.float32, np.float64]:
    print(np.finfo(dtype).min)
    print(np.finfo(dtype).max)
    print(np.finfo(dtype).eps)
````
